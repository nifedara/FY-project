```python
import pandas as pd
import spacy
import nltk
from nltk import punkt
from collections import Counter

nlp = spacy.load("en_core_web_sm")
```

```python
#wikidata = pd.read_csv(r'C:\Users\olami\Documents\tech\web projects\FinalYearProject\jsonToCsv\wikidata.csv')
wikidata = pd.read_csv(r'C:\Users\olami\Documents\tech\web projects\FinalYearProject\jsonToCsv\half_wikidata.csv')
```

```python
import csv

wikidata_location = []

for text in wikidata['text']:
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == 'GPE':
            wikidata_location.append(doc)    
```

```python
 with open('wikitext_file.csv', 'w', newline ='', encoding = 'utf-8') as f:
    w = csv.writer(f)
     for data in wikidata_location:
        w.writerow(data)
```
